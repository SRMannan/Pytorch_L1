```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns = ['id' , 'Unnamed: 32'] , inplace = True)
```

```python
df.sample(20)
```

```python
X_train , X_test , y_train , y_test = train_test_split(df.iloc[: , 1:] , df.iloc[: , 0] , test_size = 0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
print(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.fit_transform(y_test)
```

```python
y_train
```

```python
import torch
```

```python
X_train_torch = torch.from_numpy(X_train)
X_test_torch = torch.from_numpy(X_test)
y_train_torch = torch.from_numpy(y_train)
y_test_torch = torch.from_numpy(y_test)

y_train_torch = y_train_torch.type(torch.float64)
y_test_torch = y_test_torch.type(torch.float64)
```

```python
y_train_torch.shape
```

```python
X_train_torch.shape
```

#DEFINING THE MODEL

```python
class MySimpleNN():
  def __init__(self , X):

    #setting number of weights = no of features + a constant
    self.weights = torch.rand(X.shape[1] , 1 , dtype = torch.float64 , requires_grad = True)
    self.bias = torch.zeros(1 , dtype = torch.float64 , requires_grad = True)

  def forward(self , X):
    z = torch.matmul(X , self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(self, y_pred, y):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
    loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
    return loss
```

```python
#Hyperpara
learning_rate = 0.1
epochs = 25
```

Training pipeline

```python
model = MySimpleNN(X_train_torch)
model.weights

for e in range(epochs):

  #forward pass
  y_pred = model.forward(X_train_torch)

  ##loss
  loss = model.loss_function(y_pred , y_train_torch)
  print(f'FOR EPOCH {e} , LOSS : {loss.item()}')

  ###backwward pass
  loss.backward()

  ####para updates
  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad


#zero gradienr
model.weights.grad.zero_()
model.bias.grad.zero_()


```

EVALUATION

```python
with torch.no_grad():
  y_pred = model.forward(X_test_torch)
  y_pred = (y_pred > 0.5).float()

  acc =  (y_pred == y_test_torch).float().mean()
  print(f'ACC : {acc.item()}')
  print(y_pred)
```
